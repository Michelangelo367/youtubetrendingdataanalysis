Loading the required libraries

```python
import pandas as pd
import numpy as np
import scipy as sp
```

Loading the datasets into pandas dataframes

```python
usdata = pd.read_csv("../data/USvideos.csv",delimiter=',', encoding="utf-8-sig",sep='\t',header=0)
indata = pd.read_csv("../data/INvideos.csv",delimiter=',', encoding="utf-8-sig",sep='\t',header=0)
gbdata = pd.read_csv("../data/GBvideos.csv",delimiter=',', encoding="utf-8-sig",sep='\t',header=0)
cadata = pd.read_csv("../data/CAvideos.csv",delimiter=',', encoding="utf-8-sig",sep='\t',header=0)
```

```python
usdata.shape,indata.shape,gbdata.shape,cadata.shape
```

Merging the dataframes into a single dataframe

```python
data = pd.DataFrame()
```

```python
data = data.append(usdata)
data = data.append(cadata)
data = data.append(gbdata)
data = data.append(indata)
```

```python
dataframe = data
data.shape
```

```python
data.head(12)
```

Removing the thumbnail_link column as it doesn't serve our purpose

```python
data.fillna({'likes':0,'dislikes':0,'comment_count':0},inplace=True)
```

```python
data.drop(columns = 'thumbnail_link',inplace=True)
```
